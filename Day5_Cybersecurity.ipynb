```python
import pandas as pd

columns = [
    "text",
    "language",
    "thread_title",
    "spam_score",
    "replies_count",
    "participants_count",
    "likes",
    "comments",
    "shares",
    "type",
]

df = pd.read_csv("fake_news_dataset.csv", usecols=columns)

```

```python
df.head()
```

```python
df = df[df['language'] == 'english']

df = df.dropna()
```

```python
df.head()
```

```python
df = df.drop("language", axis=1)
```

```python
features = 0
feature_map ={}

def add_feature(name) :
  """ Adds a feature to the dictionary of features """

  if name not in feature_map :
    global features
    feature_map[name] = features
    features += 1


```

```python
add_feature("fake")
add_feature("real")

```

```python
def article_type(row) :
  """ Binaries target into fake or real """
  if row["type"] == "fake" :
    return feature_map["fake"]
  else :
    return feature_map["real"]
```

```python
df["type"] = df.apply(article_type, axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df)
```
